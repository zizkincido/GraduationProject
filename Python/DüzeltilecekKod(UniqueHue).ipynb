```python
import cv2
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
import csv

base_path = r"""C:\Users\Aziz"""
dataset_path = base_path+"\data_files\dataset_info_by_date_art.csv"

df = pd.read_csv(dataset_path)
df = df.reset_index(drop=True)  # make sure indexes pair with number of rows
art_HLS = pd.DataFrame(columns=['Number_of_Unique_Hues','Pixels_Belong_To_Dominant_Hue'])
for index, row in df.iterrows():
#     /content/drive/MyDrive/Database/ARC DATABASE/le_blanc/2022-02-21_13-47-58_UTC.jpg
    image = row['full_path'].replace(r'/content/drive/MyDrive',r"C:\Users\Aziz")
    img = cv2.imread(image)
    imgHLS = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    imgHLS = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
    #indices = np.where((imgHLS[:,:,1]>=0.95) & (imgHLS[:,:,1]<=0.15))
    #imgHLS = np.delete(imgHLS,indices,axis=1)
    #imgHLS = np.delete(imgHLS,indices,axis=0)
    #if(imgHLS.all()!=0):
    #    indices2 = np.where(imgHLS[:,:,2]<=0.2)
    #    imgHLS = np.delete(imgHLS,indices2,axis=1)
    #    imgHLS = np.delete(imgHLS,indices2,axis=0)
    #    if(imgHLS.all()!=0):
    unique, counts = np.unique(imgHLS[:,:,0], return_counts=True)
    dict(zip(unique, counts))
    unique_hues= pd.DataFrame(data=unique,columns=['Unique_Hues'])
    counts= pd.DataFrame(data=counts,columns=['Counts'])
    df2=pd.DataFrame(columns=['Unique_Hues','Counts'])
    df2['Unique_Hues']=unique_hues['Unique_Hues']
    df2['Counts']=counts['Counts']
    art_HLS.loc[len(art_HLS.index)] = [len(df2.index),df2['Counts'].max()]
    #    else:
    #        none=0
    #        art_HLS.loc[len(art_HLS.index)] = [none,none]
    #else:
    #    none=0
    #    art_HLS.loc[len(art_HLS.index)] = [none,none]
```

```python
art_HLS['like_rate'] = df['like_rate']
```

```python
x = art_HLS['like_rate']
y = art_HLS['Number_of_Unique_Hues'].astype(float)
my_rho = np.corrcoef(x,y)
plt.title('Correlation = %.2f' % my_rho[0,1])
plt.xlabel("Art Dataset Like Rates")
plt.ylabel("'Number of Unique Hues'")
plt.scatter(x, y)
```

```python
x = art_HLS['like_rate']
y = art_HLS['Pixels_Belong_To_Dominant_Hue'].astype(float)
my_rho = np.corrcoef(x,y)
plt.title('Correlation = %.2f' % my_rho[0,1])
plt.xlabel("Art Dataset Like Rates")
plt.ylabel("Pixels Belong To Dominant Hue")
plt.scatter(x, y)
```
