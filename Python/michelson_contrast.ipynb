```python
def michelson_contrast(file_path):
    image = cv2.imread(file_path)
    # https://stackoverflow.com/a/58826063/10138025
    # load image as YUV (or YCbCR) and select Y (intensity)
    # or convert to grayscale, which should be the same.
    # Alternately, use L (luminance) from LAB.
    Y = cv2.cvtColor(image, cv2.COLOR_BGR2YUV)[:,:,0]

    # compute min and max of Y
    min = int(np.min(Y))
    max = int(np.max(Y))

    contrast = (max-min)/(max+min)

    return float(contrast)
```

```python
dataset_path = "/content/drive/MyDrive/Database/ART DATABASE/dataset_info_by_date_art.csv"
df = pd.read_csv(dataset_path)
df['michelson_contrast'] = ""
df = df.reset_index(drop=True)  # make sure indexes pair with number of rows
df_michelson = pd.DataFrame()
for index, row in df.iterrows():
    # print(row)
    file_path = row['full_path']
    mc = michelson_contrast(file_path)
    
    df_michelson.loc[index, 'michelson_contrast'] = mc
# writing into the file
df_michelson.to_csv('/content/drive/MyDrive/Database/ART DATABASE/michelson_art.csv', index=False)
```
