```python
import cv2
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
import csv

base_path = r"""C:\Users\Aziz"""
dataset_path = base_path+"\data_files\dataset_info_by_date_art.csv"

df = pd.read_csv(dataset_path)
df = df.reset_index(drop=True)  # make sure indexes pair with number of rows
art_HSV = pd.DataFrame(columns=['Hue','Saturation'])
for index, row in df.iterrows():
#     /content/drive/MyDrive/Database/ARC DATABASE/le_blanc/2022-02-21_13-47-58_UTC.jpg
    image = row['full_path'].replace(r'/content/drive/MyDrive',r"C:\Users\Aziz")
    img = cv2.imread(image)
    img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    saturation = img_hsv[:, :, 1].mean()
    hue = img_hsv[:, :, 0].mean()
    art_HSV.loc[len(art_HSV.index)] = [hue,saturation]

    

```

```python
art_HSV['like_rate'] = df['like_rate']
#normalization
art_HSV['like_rate'] = art_HSV['like_rate']/art_HSV['like_rate'].max()
art_HSV['Hue'] = art_HSV['Hue']/art_HSV['Hue'].max()
art_HSV['Saturation'] = art_HSV['Saturation']/art_HSV['Saturation'].max()
```

```python
x = art_HSV['like_rate']
y = art_HSV['Hue']
my_rho = np.corrcoef(x,y)
plt.title('Correlation = %.2f' % my_rho[0,1])
plt.xlabel("Art Dataset Like Rates")
plt.ylabel("Average HUE")
plt.scatter(x, y)
```

```python
x = art_HSV['like_rate']
y = art_HSV['Saturation']
my_rho = np.corrcoef(x,y)
plt.title('Correlation = %.2f' % my_rho[0,1])
plt.xlabel("Art Dataset Like Rates")
plt.ylabel("Average Saturation")
plt.scatter(x, y)
```
