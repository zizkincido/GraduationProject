```python
Q1 = np.quantile(df['like_rate'],0.25)
Q2 = np.quantile(df['like_rate'],0.50)
Q3 = np.quantile(df['like_rate'],0.75)
df['like_rate_segment'] = ' '
df.shape
```

```python
df.loc[df.like_rate>=Q3, 'like_rate_segment'] = 'segment_4'
```

```python
df.loc[(df.like_rate>=Q2) & (df.like_rate<=Q3), 'like_rate_segment'] = 'segment_3'
```

```python
df.loc[(df.like_rate>=Q1) & (df.like_rate<=Q2), 'like_rate_segment'] = 'segment_2'
```

```python
df.loc[df.like_rate<=Q1, 'like_rate_segment'] = 'segment_1'
```
