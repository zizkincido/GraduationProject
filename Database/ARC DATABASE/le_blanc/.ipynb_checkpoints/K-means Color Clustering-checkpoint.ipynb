```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
```

```python
img = imread('2017-10-08_17-33-01_UTC.jpg')
```

```python
plt.figure(figsize=(12,6))
plt.imshow(img)
```

```python
img = img/255.0
```

```python
img.shape
```

```python
X=img.reshape(-1,3)
X.shape
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=2)
km.fit(X)
```

```python
img_seg=km.cluster_centers_
```

```python
img_seg=img_seg[km.labels_]
```

```python
img_seg = img_seg.reshape(X.shape)
img_seg.shape
```

```python
plt.figure(figsize=(12,10))
plt.imshow(img_seg)
```
